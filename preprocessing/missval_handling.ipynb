```python
import pandas as pd
```

# **1. AHH**
## **1.1. MissVal Detec**

```python
ahh = pd.read_csv('../data/drop-Data Faktor-Faktor IPM - AHH.csv')
ahh.drop('2016', axis=1, inplace=True)
ahh = ahh.replace(',', '.', regex=True)
ahh.head()
```

```python
# Memeriksa setiap baris apakah ada '-' di kolom tahun (dari 2010 sampai 2024)
tahun_cols = [str(tahun) for tahun in range(2010, 2025) if tahun != 2016]
mask = ahh[tahun_cols].astype(str).apply(lambda x: '-' in x.values, axis=1)

# Ambil nama kabupaten/kota yang mengandung missing value (-)
kab_kota_ahh = ahh.loc[mask, 'NAMA_KAB_KOTA']
print("Kabupaten/Kota yang mengandung missing value (-) pada tahun 2010-2024:\n", kab_kota_ahh.unique())
print("\nJumlah kabupaten/kota yang mengandung missing value (-) pada tahun 2010-2024:", len(kab_kota_ahh.unique()))
```

## **1.2. Drop Miss and kab_kota prov col**

```python
ahh_clean = ahh[~ahh['NAMA_KAB_KOTA'].isin(kab_kota_ahh)]
ahh_clean = ahh_clean.drop(columns=['NAMA_KAB_KOTA', 'PROVINSI'])
ahh_clean
```

```python
ahh_clean.to_csv('../data/ahh_clean.csv', index=False)
```
